# Cleaning

### Dropping columns

Under the Fair Work Act, there are some protected attributes. Under this Act an employer cannot lawfully discriminate or take action against an employee for any one of these reasons.

A list of these can be found here: https://www.fairwork.gov.au/employee-entitlements/protections-at-work/protection-from-discrimination-at-work

To ensure that this project and modelling is reproducable in a real-life situation, I will be dropping all columns that include protected attribute information. 

```python
data.drop(['Age', 'DistanceFromHome', 'Education', 'EducationField', 'Gender', 'MaritalStatus', 'NumCompaniesWorked', 'Over18', 'RelationshipSatisfaction', 'TotalWorkingYears', 'WorkLifeBalance'], axis = 1, inplace = True)
```

I will also be dropping the columns where the data is not useful for modelling purposes.

```python
data.drop(['EmployeeCount', 'StandardHours'], axis = 1, inplace = True)
```

And finally, will be dropping data where the same information is presented in numerous ways (e.g., monthly salary, hourly rate, etc.)

```python
data.drop(['DailyRate', 'MonthlyIncome', 'MonthlyRate'], axis = 1, inplace = True)
```
