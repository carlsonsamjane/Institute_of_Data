```python
import numpy as np

from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = loadtxt('C:/Users/carls/OneDrive/Desktop/IOD/Data/pima-indians-diabetes.data.csv', delimiter=',')
```

```python
# split into input (X) and output (y) variables
X = data[:,0:8]
y = data[:,8]
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
y.shape
```

```python
# define the keras model
model = Sequential()


#hidden layer 1
model.add(Dense(12, input_dim=8, activation='relu')) #12 neutrons, input_dim = shape of X

#hidden layer 2
model.add(Dense(8, activation='relu')) #8 neutrons


model.add(Dense(1, activation='sigmoid')) #final result/last output layer, sigmoid = binary
```

```python
model.summary()
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50, stratify=y)
```

```python
s = StandardScaler()
X_train = s.fit_transform(X_train)
X_test = s.transform(X_test)
```

```python
# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=150, batch_size=10)
#epochs = training cycles
#batch size = number of observations(rows)
```

```python
# evaluate the keras training model
loss, accuracy = model.evaluate(X_train, y_train)
print('Training Accuracy: %.2f' % (accuracy*100))
```

```python
# evaluate the keras testing model
loss, accuracy = model.evaluate(X_test, y_test)
print('Test Accuracy: %.2f' % (accuracy*100))
```

```python
# make probability predictions with the model
predictions = model.predict(X_train)
# round predictions 
rounded_train = [round(x[0]) for x in predictions]
rounded_train
```

```python
# make probability predictions with the model
predictions = model.predict(X_test)
# round predictions 
rounded_test = [round(x[0]) for x in predictions]
rounded_test
```

```python
# make class predictions with the training model
predictions = model.predict_classes(X_train)
predictions
```

```python
# make class predictions with the training model
predictions_test = model.predict_classes(X_test)
predictions_test
```

# Complete code

```python
# first neural network with keras make predictions
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
# load the dataset
dataset = loadtxt('C:/Users/carls/OneDrive/Desktop/IOD/Data/pima-indians-diabetes.data.csv', delimiter=',')
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10, verbose=0)
# make class predictions with the model
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(5):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
