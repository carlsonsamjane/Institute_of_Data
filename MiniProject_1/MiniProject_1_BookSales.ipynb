# 3. Does the price of a book impact the user ratings (i.e. does an expensive book have lower or higher ratings than a lower priced book?) 

#### The below graph displays a scatter plot with a linear regression line. This shows us that there is a negative correlation between the price of a book and its user rating (as books get more expensive, the sales numbers decrease). 

```python
y = books['User Rating']
x = books['Price']
plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.title('User Rating vs. Price')
plt.ylabel('User Rating')
plt.xlabel('Price')
plt.ylim([3.0, 5.0])
rcParams['figure.figsize'] = 10,5
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)
plt.rcParams['grid.color'] = 'gray'
plt.show()
```

#### This above graph shows us that the lower the price of a book, the higher the rating. From this, I would recommend to my company to not price a book above $20, as from this point on we can see that the user ratings will drop which could impact future sales and profits.
