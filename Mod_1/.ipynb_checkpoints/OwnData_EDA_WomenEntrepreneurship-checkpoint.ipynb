```python
import pandas as pd
import numpy as np
import seaborn as sns
import pandas_profiling
import matplotlib.pyplot as plt
```

### Importing my file: 'WomenEntrepreneurship_LaborForce'

Data that describes women in entrepreneurship and considers the country they are from, the country's level of development, it's currency, membership to the European union, etc. 

```python
women = pd.read_csv('WomenEntrepreneurship_LaborForce.csv')
women.head() #For an example of the type of data in the file.
women.tail()
```

```python
women.drop(labels=51, axis=0, index=None, columns=None, level=None, inplace=True)

#To remove NaN row (final row)
```

```python
women.drop(['Unnamed: 9', 'Unnamed: 10', 'Women_Entrepreneurship_Index', 'Entrepreneurship_Index'], axis=1)
#To remove final two 'unnamed' columns, and the columns regarding Entrepreneurship Index as this is not relevant for my business question.
```

### Exploring my data - inital EDA

```python
women.describe()
```

```python
#To begin my analysis, I will show only these details of the data that I am working with (for Part 1):

data = women[['Country', 'Female_Labor_Force_Participation_Rate']]
data.head()
```

```python
plt.scatter(x = data['Country'], y = data['Female_Labor_Force_Participation_Rate'])
plt.xticks(rotation=90)
plt.xlabel('Country') 
plt.ylabel('Female Participation Rate')
plt.tight_layout()
```

```python
plt.barh(data['Country'], data['Female_Labor_Force_Participation_Rate'])
plt.rcParams['figure.figsize']=(40,40)
plt.tick_params(axis='x', labelsize=40)
plt.tick_params(axis='y', labelsize=30)
```

## Part 1 Summary: Does each country have a different or similar level of representation of women in the workforce?
The above scatter and bar plots show us the overall spread of female employees in the workforce.
These show that most countrues sit between 40-70% female representation. But there are perhaps other ways to find some more useful information from the data. Some examples of this are detailed below:

```python
print(data.min())
```

```python
print(data.max())
```

```python
print(data.mean())
```

These summary details above show us that the range of female employees represented in the workplace varies greatly, however for the mean to show 58%, we can summise that overall representation is (on average) fairly balanced, not accounting for the extreme values.

## Part 2: How does gender balance/imbalance in the workforce impact inflation?

```python
inflation = women[['Female_Labor_Force_Participation_Rate', 'Inflation_rate']]
inflation.head()
```

```python
sns.pairplot(inflation)
```

```python
plt.scatter(x = women['Female_Labor_Force_Participation_Rate'], y = women['Inflation_rate'])
plt.xlabel('Female Participation in workforce')
plt.ylabel('Inflation')
```

### Part 2 Summary: How does gender balance/imbalance in the workforce impact inflation?

Inflation is not impacted by gender balance or imbalance in the workforce.

In today's environment, and with the current gender diversity intiatives in place, this argument supports a balance of gender in the workplace with regard to inflation; there is no correlation between gender balance and inflation (both genders have equal change of impacting inflation).
